```python
import matplotlib.pyplot as plt
import numpy as np
from math import sqrt
from Sum_of_squares import degeneracy

n = 3
num_max = 10000

x = []
y = []
w = []
for N in range(num_max):
    d = degeneracy(N,n)
    if d[0]:
        y.append(d[1])
        x.append(N)
        w.append(0.5*sqrt(N))

y = np.array(y)
x = np.array(x)
```

```python


fig, ax = plt.subplots()
ax.scatter(x,y, color='blue', marker=".")
ax.plot(x,w, color = 'red', label = '0.5* sqrt(N)')
ax.set_title("degen vs. sum for 3 summands")
ax.set_xlabel("value of sum N")
ax.set_ylabel("degeneracy")
ax.legend()
```
