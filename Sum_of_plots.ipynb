```python
import matplotlib.pyplot as plt
import numpy as np
from Sum_of_squares import degeneracy

n = 3
num_max = 10000
x = np.arange(1,num_max+1,1)

y = []
for j in range(num_max):
    y.append(degeneracy(j,n)[1])
yp = np.array(y)
print(np.size(y))
print(np.size(x))


fig, ax = plt.subplots()
ax.scatter(x,yp, color='blue', marker=".")
ax.set_title("degen vs. sum")
ax.set_xlabel("value of sum")
ax.set_ylabel("degeneracy")
```
