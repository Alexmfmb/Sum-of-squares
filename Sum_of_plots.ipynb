```python
import matplotlib.pyplot as plt
import numpy as np
from math import sqrt
from Sum_of_squares import degeneracy

```

## Long calculation

```python
n = 3

num_min = 1
num_max = 100

deg_array = []

for N in range(num_min,num_max):
    deg_array.append(degeneracy(N,n))
```

## Create data for plot

```python
x = []
y = []
x_max = []
y_max = []

for i in range(len(deg_array)):
    if deg_array[i][0]:
        y.append(deg_array[i][1])
        x.append(num_min + i)

    if y_max and deg_array[i][0] and deg_array[i][1] > max(y_max):
        y_max.append(deg_array[i][1])
        x_max.append(num_min + i)
```

## Creating plot

```python
w = []
xw = []
for N in range(num_min,num_max):
    xw.append(N)
    w.append(1/2.5*sqrt(N))


fig, ax = plt.subplots()
ax.scatter(x,y, color='blue', marker=".")
#ax.scatter(x_max,y_max, color='green', marker=".")
ax.plot(xw,w, color = 'red', label = '2.5 * sqrt(N)')
ax.set_title("degen vs. sum for 3 summands")
ax.set_xlabel("value of sum N")
ax.set_ylabel("degeneracy")
ax.legend()
```
